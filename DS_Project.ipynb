# Data Preprocessing

```python
#importing Libraries that might come in handy later
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#Label Encoder is a library which assigns a unique numerical value to every string in a list of strings.
#If a word is present no. of times it is given the same number all the time to maintain data accuracy and veracity
Encoder = LabelEncoder()
venues = Encoder.fit_transform(df['Venue'])
print(venues)
```

```python
#In this loop all the values are updated with their encoded number
for i in range(2016):
    df['Venue'][i] = venues[i]
```

This Process of the above 2 cells is used below for all the features which will be used in the model

```python
#Using the above process for feature Toss Winner
toss = Encoder.fit_transform(df['Toss_winner'])
print(toss)
for i in range(2016):
    df['Toss_winner'][i] = toss[i]
```

```python
#Using the above process for feature Toss Decision
tossD = Encoder.fit_transform(df['Toss_decision'])
print(tossD)
for i in range(2016):
    df['Toss_decision'][i] = tossD[i]
```

```python
#Using the above process for feature Winner
winner = Encoder.fit_transform(df['Winner'])
print(winner)
for i in range(2016):
    df['Winner'][i] = winner[i]
```

# Data Model

```python
#Importing SciKit learning model selection library and importing train_test_split to split dattaset into training
# and testing to train and test the models 
from sklearn.model_selection import train_test_split
# Defining features list with all the variables which will be used in the models
players = ['Players_1','Players_2','Players_3','Players_4','Players_5','Players_6','Players_7','Players_8','Players_9','Players_10','Players_11']
```

# Importing SVM data model as the first model

```python
#Importing the required libraries
from sklearn import svm
from sklearn.metrics import accuracy_score
#Testing array will contain all the 11 players Y testing list
testing = []
#acr is a variable to store accuracy after every player so average can be calculated
acr = 0
```

```python
def player_pred(df,player_number,acr) :
    features = ['Team1','Team2','Winner','Toss_decision','Venue']
    X = df.loc[:, features].values
    # Split dataset into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, df[players[player_number]], test_size=0.2,random_state=109)
    clf = svm.SVC(kernel='linear',probability=True)
    # Applying the algorith to training data set
    clf.fit(X_train, y_train)
    #adding testing data set to testing array for calculating individual player accuracy
    testing.append(y_test)
    # Predicting the players based on the algorithm
    predicted = clf.predict(X_test)
    #adding algorithm accuracy for predicting this player 
    acr += accuracy_score(y_test, predicted)
    #aranging output predicted players accourding to their probability or likelyness to play
    results = clf.predict_proba(X_test)[0]
    prob_per_class_dictionary = dict(zip(clf.classes_, results))
    res = list(map(lambda x: x[0], sorted(zip(clf.classes_, results), key=lambda x: x[1], reverse=True)))
    #Since sometimes not all 11 players get to play so many cells have - to remove that and take only players and
    # return both the player and its accuracy
    if res[0] != '-':
        return [res[0],acr]
    else:
        return [res[1],acr]
```

```python
# List which will contain the 11 players 
playing_11 = []
```

```python
# Loop to run the method 11 times for the 11 players
for i in range (11):
    a = player_pred(df,i,acr)
    playing_11.append(a[0])
    acr = a[1]
#printing the playing 11
print(playing_11)
```

```python
#Printing the accuracy in percentage
print(acr*100, "% is the accuracy of the model")
```

```python
#Custom method to find accuracy of individual players
def accuracy(player_number):
    count = 0
    for j in range (11):
        for i in testing[j].values:
            if (i == playing_11[player_number]):
                count += 1
    return count
```

```python
# Loop to find the accuracy of only the 11 players
sum = 0
for i in range (11):
    sum += accuracy(i)
print((sum/11)/len(testing[0].values)*100)
```

```python
# This method finds the dataset where 2 teams are playing do that a playing 11 can be found when these 2 teams play
def country(country,country2):
    df_country1 = df[(df['Team1']==country) & (df["Team2"]==country2)]
    df_country2 = df[(df['Team1']==country2) & (df["Team2"]==country)]
    frames = [df_country1, df_country2]
    df_country = pd.concat(frames)
    return df_country
```
